```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel("Book1.xlsx")
dataset.head(8)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.distplot(dataset["Col"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create an instance of MinMaxScaler
scaler = MinMaxScaler()

# Fit the scaler to your data
scaler.fit(dataset[["Col"]])
```

```python
scaler.transform(dataset[["Col"]])
```

```python
dataset["Col min"] = scaler.transform(dataset[["Col"]])
```

```python
dataset.head(7)
```

```python
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["Col"])

plt.subplot(1,2,2)
plt.title("after")
sns.distplot(dataset["Col min"])
plt.show()
```
