```python
market=[]
for i in range(0,dataset.shape[0]):
    cus=[]
    for j in dataset.columns:
        if type(dataset[j][i])==str:
          cus.append(dataset[j][i])
    market.append(cus)      
```

```python
l=[]
for i in market:
    for j in i:
        l.append(j)
```

```python
p=collections.Counter(l)
```

```python
p.keys()
```

```python
p.values()
```

```python
d= {"item name":p.keys(),"values":p.values()}
```

```python
pd.DataFrame(d)
```

```python
pd.DataFrame(d).sort_values(by=["values"],ascending=False)
```

```python
tr=TransactionEncoder()
tr.fit(market)
```

```python
df=pd.DataFrame(tr.transform(market),columns=tr.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(df,min_support=0.07,use_colnames=True,max_len=3)
```

```python
apriori(df,min_support=0.07,use_colnames=True,max_len=3).sort_values(by=["support"])
```
