```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset= pd.read_excel("Book3.xlsx")
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x=dataset.iloc[:,:-1]
y=dataset["place"]
```

```python
sc= StandardScaler()
sc.fit(x)
x=pd.DataFrame(sc.transform(x),columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn= KNeighborsClassifier(n_neighbors=9)
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)*100
```

```python
knn.score(x_train,y_train)*100
```

```python
for i in range(1,20):
    dt= KNeighborsClassifier(n_neighbors=i)
    dt.fit(x_train,y_train)
    print(i,dt.score(x_train,y_train)*100, dt.score(x_test,y_test)*100)
```

```python
#5.6	1.980	1
```

```python
dt.predict([[5.6, 1.980]])
```

```python
#6.7	3.250	0
```

```python
dt.predict([[6.7, 3.250]])
```

```python
#8.9	2.815	2
```

```python
dt.predict([[8.9, 2.815]])
```

```python
sns.scatterplot(x="cgpa",y="score", data=dataset, hue="place")
plt.show()
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=dt)
plt.show()
```
