```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel(r"Book1.xlsx")
dataset.head(6)
```

```python
dataset.isnull().sum()
```

```python
dataset["Item"] = pd.to_numeric(dataset["Item"], errors="coerce")
```

```python
Item_mean = dataset['Item'].mean()
```

```python
# dataset["Item"].fillna(dataset["Item"].mean(), inplace=True)
dataset['Item'] = dataset['Item'].fillna(Item_mean)
```

```python
dataset["Item"].fillna(dataset["Item"].mean(), inplace=True)
```

```python
dataset.head(8)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["Item"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
ss.fit(dataset[["Item"]])
```

```python
ss.transform(dataset[["Item"]])
```

```python
dataset["Item_ss"] = pd.DataFrame(ss.transform(dataset[["Item"]]), columns=["x"])
```

```python
dataset.head(7)
```

```python
dataset.describe()
```

```python
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["Item_ss"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["Item"])

plt.show()
```
