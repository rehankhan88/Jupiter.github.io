```python
import pandas as pd

# Assuming 'dataset' is your DataFrame

# Inspect the Invoice column
print(dataset['Invoice '])

# Convert the Invoice column to numeric, coercing errors to NaN
dataset['Invoice '] = pd.to_numeric(dataset['Invoice '], errors='coerce')

# Drop rows with NaNs in the Invoice column
dataset = dataset.dropna(subset=['Invoice '])

# Recalculate the mean and standard deviation
min_range = dataset['Invoice '].mean() - (3 * dataset['Invoice '].std())
max_range = dataset['Invoice '].mean() + (3 * dataset['Invoice '].std())

print(min_range, max_range)
```

```python
new_dataset = dataset[dataset["Invoice "]<=max_range]
```

```python
sns.boxplot(x="Invoice ", data=new_dataset)
```
