```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import seaborn as sns
```

```python
pd.set_option("display.max_rows",550)
```

```python
x,y=make_moons(n_samples=500,noise=0.2)
```

```python
df={"x1":x[:,0],"x2":x[:,1],"y":y}
```

```python
dataset=pd.DataFrame(df)
```

```python
dataset.head(5)
```

```python
sns.scatterplot(x="x1",y="x2",data=dataset,hue="y")
plt.show()
```

```python
xa=dataset.iloc[:,:-1]
ya=dataset["y"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(xa,ya,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
dt.score(x_train,y_train)*100, dt.score(x_test,y_test)*100
```

```python
sv=SVC()
sv.fit(x_train,y_train)
```

```python
sv.score(x_train,y_train)*100, sv.score(x_test,y_test)*100
```

```python
g=GaussianNB()
g.fit(x_train,y_train)
```

```python
g.score(x_train,y_train)*100, g.score(x_test,y_test)*100
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
li=[("dt1",DecisionTreeClassifier()),("sv1",SVC()),("g1",GaussianNB())]
```

```python
vc=VotingClassifier(li,weights=[10,4,7])
vc.fit(x_train,y_train)
```

```python
vc.score(x_train,y_train)*100, vc.score(x_test,y_test)*100
```

```python
prd= {"dt":dt.predict(x_test),"svm":sv.predict(x_test),"g":g.predict(x_test),"vc":vc.predict(x_test)}
```
