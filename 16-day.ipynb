```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
dataset = pd.read_excel("Book2.xlsx")
dataset.head(7)
```

```python
x = dataset.iloc[:,:-1]
y = dataset["oc"]
```

```python
x.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
fs = SequentialFeatureSelector(lr,k_features=3,forward=True)
fs.fit(x,y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```
