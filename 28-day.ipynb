```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset= pd.read_excel("Book3.xlsx")
dataset.head(5)
```

```python
sns.kdeplot(data=dataset["cgpa"])
plt.show()
```

```python
sns.scatterplot(x="cgpa",y="score",data=dataset,hue="place")
plt.show()
```

```python
x= dataset.iloc[:,:-1]
y= dataset["place"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
lr = GaussianNB()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.score(x_train,y_train)*100
```

```python
mb= MultinomialNB()
mb.fit(x_train,y_train)
```

```python
mb.score(x_test,y_test)*100, mb.score(x_train,y_train)*100
```

```python
bnb= BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)*100, mb.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr)
plt.show()
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mb)
plt.show()
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```
