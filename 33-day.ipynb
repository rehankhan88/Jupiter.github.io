```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset= pd.read_excel("Book8.xlsx")
dataset.head(3)
```

```python
x=dataset.iloc[:,:-1]
y=dataset["salary"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor(criterion="squared_error",splitter="best",max_depth=10)
dt.fit(x_train,y_train)
```

```python
dt.score(x_train,y_train)*100 , dt.score(x_test,y_test)*100
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
df={"criterion":["squared_error", "friedman_mse", "adsolute_error", "poisson"],
    "splitter":["best", "random"],
    "max_depth":[i for i in range(2,20)]
    }
```

```python
gd=GridSearchCV(DecisionTreeRegressor(),param_grid=df)
gd.fit(x_train,y_train)
```

```python
gd.best_params_
```

```python
gd.best_score_*100
```

```python
rd= RandomizedSearchCV(DecisionTreeRegressor(),param_distributions=df,n_iter=20)
rd.fit(x_train,y_train)
```

```python
rd.best_params_
```

```python
rd.best_score_*100
```
